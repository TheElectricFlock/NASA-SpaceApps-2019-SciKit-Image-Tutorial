#### Images as Numpy arrays

Scikit image is built to be interoperable with Numpy and Scipy. In fact...pictures in scikit-image are actually Numpy arrays! Lets have a look!

```python
import numpy as np
from matplotlib import pyplot as plt

random_static = np.random.random([28, 28])
```

Here we have created a 2D numpy array with random values ranging from 0 to 1, What will this look like?

```python
plt.imshow(random_static, cmap='gray')
plt.colorbar()
```

He we see a randomly generated 2D Numpy array diplayed as an image. Notice the colourbar, and how each array element (or pixel in this context) has a value from 0 to 1. With zero being the darkest and white being the brightest. Lets unpack this a tad.

```python
print('Type: ', type(random_static))
print('Datatype: ', random_static.dtype)
print('Shape: ', random_static.shape)
```

So here we can see our random image is a numpy ndarray, it's pixels are 64 bit floats, and it's shape is 28 by 28 pixels. Now with this example for every pixel there is one value going from 0 to a given maximum value. We usually call this kind of colourspace as "greyscale".

As this image is a numpy array, there is no reason we can't treat it as such. What do we think is going to happen when we call this block of code below?

```python
random_static[0:14,0:14] = 1
plt.imshow(random_static, cmap='gray')
```

We can see here that the array column and row values for index 0 to 14 have all been set to 1, which is the brightest value.

#### A quick note about pixel values:

In literature greyscale pixel values are typicall represented as intgers from:  
0 - 1 where 0 is black and 1 is white  
0 - 255 where 0 is black and 255 is white

Ultimately it is the data type of the Numpy array that determined the pixel ranges, but Scikit-image supports both conventions (and more).

#### Visualising colour channels  
So as we've seen greyscale images can be represented as arrays. But how do we represent colours in images?

```python
cat = data.chelsea()
plt.imshow(cat)
```

Here we see a cat...IN COLOUR! but how do arrays represent colour?

```python
##Lets have a look at the very first pixel
cat[0][0]
```

Here we see the pixel is an array of 3 values, and a datatype. This is because this colour channel is made up of values consisting of red intensity (where 0 is black and 255 is red), green intensity (where 0 is black and 255 is green), and blue intensity (where 0 is black and 255 is blue).

There are many different types of colour channels but for now we will stick in the RGB domain.
